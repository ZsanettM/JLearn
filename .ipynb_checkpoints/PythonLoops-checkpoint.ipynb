<a id="while"><strong>while loop example:</strong></a>

```python
x = 5
while x != 0:
    print(x)
    x = x - 1
    print("Wow, we've counted x down, and now it equals", x)
print ("And now the loop has ended.")
```

<a id="for"><strong>for loop example:</strong></a>

```python
word = input("Please enter a word: ") #get user input

for letter in word:
    if (letter == word[0]):
        print("The first letter is: \t",letter)
    elif (letter == word[-1]):
        print("The last letter is: \t",letter)
    else:
        print("The next letter is: \t",letter)
```
